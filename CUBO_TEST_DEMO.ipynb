```python
from tqdm import tqdm
import numpy as np
import dask.array as da
data_lake = pd.DataFrame()
for ii in tqdm(range(len(merged_df)), desc="Processing Lakes"):
  tmp_df = pd.DataFrame()
  c = merged_df.iloc[ii,0]
  center_lat = merged_df.iloc[ii,1]
  center_lon = merged_df.iloc[ii,2]
  start_time= merged_df.iloc[ii,-2]
  end_time= merged_df.iloc[ii,-1]

  #ERA5 IRR
  da_IRR = cubo.create(
    lat=center_lat,
    lon=center_lon,
    collection=ERA5_key[0], # ID of the GEE collection Landsat_sensor[0]
    bands=["surface_net_solar_radiation_sum"], # Bands to retrieve
    start_date=start_time,
    end_date=end_time, # End date of the cube (remember in GEE this date is not included)
    edge_size=4,
    resolution=30,
    gee=True # Set to True
  )

  irr = da_IRR/86400
  irr = da.mean(irr, axis=(2, 3), keepdims=True)
  irr_array = irr.compute()
  tmp_df['time'] = da_IRR.time.values
  tmp_df['lakename'] = c
  tmp_df['irr'] = pd.DataFrame(irr_array.reshape(-1,1))
  data_lake= pd.concat([data_lake, tmp_df], ignore_index=True)
```
